```python
Smile_string=df['SMILES']
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
from rdkit.Chem import Fragments
```

```python
Smile_list=list(Smile_string)
```

```python
print(Smile_list)
```

```python
Mol_list=[]
for smiles in Smile_list:
    Mol=Chem.MolFromSmiles(smiles)
    Mol_list.append(Mol)
```

```python
print('there are', len(Smile_list),'chemicals in the dataset')
```

```python
print(Mol_list)
```

```python
K=Mol_list[:200]
#this shows the image for the first 200 chemicals.
```

```python
img=Draw.MolsToGridImage(K,molsPerRow=3)
img
```

```python
Mw_list=[]
M=Mol_list
```

```python
for chem in M:
    Mw=Descriptors.MolWt(chem)
    Mw_list.append(Mw)
print(Mw_list)
print(len(Mw_list))
```

```python
C=pd.DataFrame(Mw_list)
C
```

```python
df['MOLECULAR WEIGHT']=C
```

```python
df.head()
```

```python
Ester_list=[]
for esters in Mol_list:
    E=Fragments.fr_ester(esters)
    Ester_list.append(E)
len(Ester_list)
Est=pd.DataFrame(Ester_list)
Est
df['ESTER PRESENT']=Est
df.head()
```
